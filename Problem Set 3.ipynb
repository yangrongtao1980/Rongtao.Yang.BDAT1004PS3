# Question 1. Occupations

```python
# Step 1-3
import pandas as pd
users = pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users.head()
```

```python
# Step4 Discover what is the mean age per occupation.
users.groupby('occupation')['age'].mean()
```

```python
# Step6 For each occupation, calculate the minimum and maximum ages.
users.groupby('occupation')['age'].agg(['min', 'max'])
```

```python
# Step7 For each combination of occupation and sex, calculate the mean age.
users.groupby(['occupation', 'gender'])['age'].mean()
```

```python
# Step8 For each occupation present the percentage of women and men.
users.groupby(['occupation', 'gender']).size() / users.groupby('occupation').size()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
data.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per Capita')
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per Capita')
plt.legend()

plt.xlabel('Year')
plt.ylabel('Rate per 1000 People')
plt.title('Marriage and Divorce Rates per Capita in the U.S. (1867-2014)')

plt.show()
```

# Question 7

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
data.head()
```

```python
years = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(years)]

plt.figure(figsize=(10, 6))
plt.bar(filtered_data['Year'] - 2, filtered_data['Marriages_per_1000'], width=4, label='Marriages per Capita')
plt.bar(filtered_data['Year'] + 2, filtered_data['Divorces_per_1000'], width=4, label='Divorces per Capita')

plt.xlabel('Year')
plt.ylabel('Rate per 1000 People')
plt.title('Comparison of Marriage and Divorce Rates per Capita (1900, 1950, 2000)')
plt.xticks(years)

plt.legend()

plt.show()
```

# Question 8

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("actor_kill_counts.csv")
data.head()
```

```python
sorted_data = data.sort_values(by='Count', ascending=True)
print(sorted_data)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(sorted_data['Actor'], sorted_data['Count'], color='skyblue')
for index, value in enumerate(sorted_data['Count']):
    plt.text(value + 1, index, str(value), va='center')

plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood by Kill Count')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("roman-emperor-reigns.csv")
data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("roman-emperor-reigns.csv")

assassination_count = data[data['Cause_of_Death'] == 'Assassinated'].shape[0]
other_causes_count = data[data['Cause_of_Death'] != 'Assassinated'].shape[0]

assassination_count = data[data['Cause_of_Death'] == 'Assassinated'].shape[0]
other_causes_count = data[data['Cause_of_Death'] != 'Assassinated'].shape[0]

sizes = [assassination_count, other_causes_count]
labels = ['Assassinated', 'Other Causes']
colors = ['green', 'blue']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  
plt.title("Fraction of Roman Emperors Assassinated")

plt.show()
```
